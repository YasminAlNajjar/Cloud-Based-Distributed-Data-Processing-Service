```python
# قراءة الملف من Unity Catalog Volume مع تحديد الفاصل ;
df = spark.read.csv(
    "/Volumes/workspace/default/project_data/students.csv",
    header=True,
    inferSchema=True,
    sep=";"   # هنا الفاصل الصحيح
)

df.show(5)
df.printSchema()
```

```python

# df.count() → يعطي عدد الصفوف
# len(df.columns) → يعطي عدد الأعمدة

print("Number of rows:", df.count())
print("Number of columns:", len(df.columns))
```

```python
from  pyspark.sql.functions import col, sum

# لكل عمود، نتحقق من القيم الفارغة (Null)
# sum(col(c).isNull().cast("int")) → يحسب عدد القيم Null
null_counts = df.select([sum(col(c).isNull().cast("int")).alias(c) for c in df.columns])

# عرض النتائج
null_counts.show()
```

```python
# نحدد الأعمدة الرقمية فقط
numeric_cols = ["age", "Medu", "Fedu", "traveltime", "studytime", 
                "failures", "absences", "G1", "G2", "G3"]

# df.select(...).describe() → يعطي الإحصائيات الأساسية لكل عمود رقمي
df.select(numeric_cols).describe().show()
```
